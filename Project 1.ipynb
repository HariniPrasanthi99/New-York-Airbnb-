**Task.4 Data Cleaning**

```python
data.isnull().sum()
```

```python
#dropping all missing values rows
data.dropna(inplace= True)
```

```python
#checking if there are any null values after dropna
data.isnull().sum()
```

```python
data.shape
```

```python
#dealing with duplicate rows
data.duplicated().sum()
```

```python
data[data.duplicated()]
```

```python
#deleting all duplicate rows
# data[data.duplicated()]

data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
data[data.duplicated()]
```

```python
#type casting
#changing data types
data.dtypes

#to change data type of id from float to object

data['id'] = data['id'].astype(object)
data.dtypes

#to change data type of host_id from int to object
data['host_id']= data['host_id'].astype(object)
data.dtypes
```

**Task.5 Data Analysis**

Univeriate Analysis

```python
#Identifying outliers in price

df= data[data['price']<1500]
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=df, x='availability_365')
plt.title('Availability_365 Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
df.dtypes
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

**Feature Engineering**


price of the room depends on the number of beds, create a new column to find the price per bed

```python
## Price per bed

df['price per bed']= df['price']/df['beds']
df.head()
```

```python
#Average price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

**Biveriate Analysis**

One variable dependency on another variable

```python
df.columns
```

```python
#Price dependency on neighbourhood
sns.barplot(data=df, x='neighbourhood_group', y= 'price', hue= 'room_type')
```

```python
#Number of reviews and Price relationship
plt.figure(figsize=(8,5))
plt.title("Locality and Review Dependency")
sns.scatterplot(data=df, x='number_of_reviews', y= 'price', hue= 'neighbourhood_group')
plt.show()
```
